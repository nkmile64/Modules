# Getting and Setting Options

To change an option call `.set_option("option regex", new_value)`

```python
# For example, this option simulates interactive mode for purposes of debugging.
pd.get_option("mode.sim_interactive")
```

```python
pd.set_option("mode.sim_interactive", True)
```

```python
pd.get_option("mode.sim_interactive")
```

To revert to the option's default value use `.reset_option()`

```python
pd.get_option("display.max_rows")
```

```python
pd.reset_option("display.max_rows")
```

```python
pd.get_option("display.max_rows")
```

To execute code with a set of given values, and then restore the original values:

```python
with pd.option_context("display.max_rows", 10, "display.max_columns", 5):
    print(pd.get_option("display.max_rows"))
    print(pd.get_option("display.max_columns"))
```

```python
# When the above block of code finishes, values are restored to defaults:
print(pd.get_option("display.max_rows"))
print(pd.get_option("display.max_columns"))
```

# Frequently Used Options

```python
pd.reset_option("all")

df = pd.DataFrame(np.random.randn(7, 2))
pd.set_option("display.max_rows", 7)
df
```

```python
pd.set_option("display.max_rows", 4)
df
```

```python
pd.reset_option("display.max_rows")
```

```python
pd.set_option("display.max_rows", 8)
pd.set_option("display.min_rows", 4)

# below max_rows -> all rows shown
df = pd.DataFrame(np.random.randn(7, 2))
df
```

```python
# above max_rows -> only min_rows(4) rows shown
df = pd.DataFrame(np.random.randn(9, 2))
df
```

```python
pd.reset_option("display.max_rows")
pd.reset_option("display.min_rows")
```

```python
df = pd.DataFrame(np.random.randn(5, 15))
pd.set_option("expand_frame_repr", True)
df
```

```python
pd.set_option("expand_frame_repr", False)
df
```
