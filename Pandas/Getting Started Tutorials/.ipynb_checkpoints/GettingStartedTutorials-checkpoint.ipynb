# Create plots in pandas

```python
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# index_col = 0     : define the first(0th) column as index of the data frame. 
# parse_dates = True: convert the dates in the column to Timestamp objects. 
#                     (Timestamp is the pandas equivalent of the Python Datetime object) 
air_quality = pd.read_csv("data/air_quality_no2.csv", index_col = 0, parse_dates = True)
```

```python
air_quality.head()
```

```python
# A quick visual check of the data 
air_quality.plot()
plt.show()
```

```python
# Visually compare the NO2 values in London vs Paris. 
air_quality.plot.scatter(x = "station_london", y = "station_paris", alpha = 0.5)
plt.show()
```

```python
# An overview of the available plot methods
[
    method_name 
    for method_name in dir(air_quality.plot) 
    if not method_name.startswith("_")
]
```

```python
# Visualize a boxplot 
air_quality.plot.box() 
plt.show()
```

```python
# Plot each of the columns in a separate subplot 
air_quality.plot.area(figsize=(12,4), subplots=True)
plt.show()
```

```python
# Customize, extend & save the resulting plot 

# Create an empty Matplotlib Figure and Axes.
fig, axs = plt.subplots(figsize = (12, 4))
# Use pandas to put the area plot on the prepared Figure/Ages. 
air_quality.plot.area(ax = axs)
# Do any Matplotlib customization you like. 
axs.set_ylabel("NO$_2$ concentration")
# Save the Figure/Axes using the existing Matplotlib method. 
fig.savefig("no2_concentrations.png")

# Display the plot. 
plt.show()
```

# How to create new columns derived from existing columns

```python
import pandas as pd 
```

```python
air_quality = pd.read_csv("data/air_quality_no2.csv", index_col = 0, parse_dates = True)
```

```python
air_quality.head() 
```

```python
# Express the NO2 concentration of London in mg/m3
# (the conversion factor is 1.882)
air_quality["london_mg_per_cubic"] = air_quality["station_london"] * 1.882
```

```python
air_quality.head()
```

```python
# New column with the ration of the values in Paris over Antwerp. 
air_quality["ratio_paris_antwerp"] = (air_quality["station_paris"] / air_quality["station_antwerp"])
```

```python
air_quality.head()
```

```python
# Rename the data columns to the station identifiers 
air_quality_renamed = air_quality.rename(columns = {
    "station_antwerp": "BETR801", 
    "station_paris": "FR04014", 
    "station_london": "London Westminster"
})
```

```python
air_quality_renamed.head()
```

```python
# Convert the column names to lower case letters. 
air_quality_renamed = air_quality_renamed.rename(columns = str.lower)
```

```python
air_quality_renamed.head()
```

# How to calculate Summary Statistics

```python
import pandas as pd 
```

```python
titanic = pd.read_csv("data/titanic.csv")
```

```python
titanic.head()
```

## Aggregating statistics

```python
# Average age of the Titanic passengers
titanic["Age"].mean()
```

```python
# Median age and ticket fare price
titanic[["Age", "Fare"]].median()
```

```python
# Pre-defined statistics
titanic[["Age", "Fare"]].describe()
```

```python
# User-defined combinations of aggregating statistics
titanic.agg(
    {
        "Age": ["min", "max", "median", "skew"], 
        "Fare": ["min", "max", "median", "mean"]
    } 
)
```

## Aggregating statistics grouped by Category

```python
# Average age for male versus female passengers
titanic[["Sex", "Age"]].groupby("Sex").mean()
```

```python
# Selection of columns is supported on the grouped data
# as well 
titanic.groupby("Sex")["Age"].mean()
```

```python
# Mean ticket fare price for each sex and cabin combinations. 
titanic.groupby(["Sex", "Pclass"])["Fare"].mean()
```

## Count number of records by category

```python
# Number of passengers in each cabin class. 
titanic.groupby("Pclass")["Pclass"].count()
```

```python
# Same but use a shortcut 
titanic["Pclass"].value_counts()
```

## Long to wide table format

```python
# Filter for NO2 data only
no2 = air_quality[air_quality["parameter"] == "no2"]
```

```python
no2.head()
```

```python
# Use 2 measurements (head) for each location (groupby) 
no2_subset = no2.sort_index().groupby(["location"]).head(2) 
```

```python
no2_subset
```

```python
# Values of the three stations as separate columns side by side. 
no2_subset.pivot(columns = "location", values = "value")
```

```python
plt.figure(figsize = (20, 8))
no2.pivot(columns = "location", values = "value").plot()
plt.show()
```

## Pivot table

```python
# Mean concentrations for NO2 and PM2 in each of the stations. 
air_quality.pivot_table(values = "value", index = "location", columns = "parameter", aggfunc = "mean")
```

```python
air_quality.pivot_table(
    values = "value",
    index = "location",
    columns = "parameter",
    aggfunc = "mean",
    margins = True) 
```

## Wide to long format

```python
# Add a new index to the data frame 
no2_pivoted = no2.pivot(columns = "location", values = "value").reset_index()
```

```python
no2_pivoted.head()
```

```python
# All air NO2 measurements in a single column. 
# Use the melt() method to convert the data table from 
# wide format(many columns) --> long format. 
#
# All columns NOT mentioned in id_vars will be melted in 
# two columns: 
# (a) The first will contain the column header names (BETR801,
#     FR04014, London Westminster)
# (b) The second contains the values themselves. 
no_2 = no2_pivoted.melt(
    id_vars = "date.utc",
    value_vars = ["BETR801", "FR04014", "London Westminster"], 
    value_name = "NO_2",
    var_name = "id_location"
) 
```

```python
no_2.head()
```

# Combine data from multiple tables

```python
import pandas as pd 
```

```python
air_quality_no2 = pd.read_csv("data/air_quality_no2_long.csv", parse_dates = True)
```

```python
air_quality_no2 = air_quality_no2[["date.utc", "location", "parameter", "value"]]
```

```python
air_quality_no2.head()
```

```python
air_quality_pm25 = pd.read_csv("data/air_quality_pm25_long.csv", parse_dates = True)
```

```python
air_quality_pm25 = air_quality_pm25[["date.utc", "location", "parameter", "value"]]
```

```python
air_quality_pm25.head()
```

## Concatenating objects

```python
# Combine the measurements of NO2 and PM25 two tables, 
# with a similar structure, in a single table. 
# 
# Concatenation along axis 0 combines the ROWS of the input tables. 
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis = 0)
```

```python
air_quality.head()
```

```python
# Check the shape of original and concatenated tables. 
print("Shape of  the ''air_quality_pm25'' table: ", air_quality_pm25.shape)
```

```python
print("Shape of  the ''air_quality_no2'' table: ", air_quality_no2.shape)
```

```python
print("Shape of the resulting \"air_quality\" table: ", air_quality.shape)
```

```python
air_quality = air_quality.sort_values("date.utc")
```

```python
air_quality.head()
```

```python
# Using the keys argument, provides an additional (hierarchical) row index 
air_quality_ = pd.concat([air_quality_pm25, air_quality_no2], keys = ["PM25", "NO2"])
```

```python
air_quality_.head()
```

# How to handle time series data with ease

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv("data/air_quality_no2_long.csv")
```

```python
air_quality = air_quality.rename(columns = {"date.utc": "datetime"})
```

```python
air_quality.head()
```

```python
# Return the unique values of the "city" column. 
air_quality.city.unique()
```

## Using pandas datetime properties

```python
# Convert the values in "datetime" column (stored in the   
# data frame as strings) to pandas datetime objects.
air_quality["datetime"] = pd.to_datetime(air_quality["datetime"])
```

```python
air_quality.datetime
```

```python
# Find the start and end date of the time series data set 
air_quality.datetime.min(), air_quality.datetime.max()
```

```python
# Calculate the duration of our time series
air_quality.datetime.max() - air_quality.datetime.min()
```

```python
# Add a new column to the dataframe, containing only the month of the date 
# NOTE: Now that the datetime column contains Timestamp objects, we can 
#       use the dt accessor to return month, year etc. 
air_quality["month"] = air_quality["datetime"].dt.month
```

```python
air_quality.head()
```

```python
# Return the average NO2 concentration: 
#     for each day of the week, 
#     for each of the measurement locations. 
air_quality.groupby([air_quality.datetime.dt.weekday, air_quality.location]).value.mean()
```

```python
# Plot the average NO2 value for each hour of the day i.e. the NO2 pattern
# during the day of all stations combined. 
fig, axs = plt.subplots(figsize = (12, 4))

(air_quality
    .groupby(air_quality["datetime"].dt.hour)["value"]
    .mean()
    .plot(kind = "bar", rot = 0, ax = axs))

plt.xlabel("Hour of the day")
plt.ylabel("$NO_2 (μg/m^3)$")

plt.show()
```

## Datetime as index

```python
# Reshape the data table with each of the station locations, as a separate column 
no_2 = air_quality.pivot(index = "datetime", columns = "location", values = "value")
no_2.head()
```

```python
# With a datetime index, we don't need the dt accessor to get the time series properties
no_2.index.year, no_2.index.weekday
```

```python
# Plot the NO2 values in the different stations from 20 May till 21 May 
no_2["2019-05-20":"2019-05-21"].plot()
```

## Resample a time series to another frequency

```python
# Aggregate the current hourly time series values to 
# the maximum monthly (ME=MonthEnd) value in each of the stations. 
monthly_max = no_2.resample("ME").max()
```

```python
monthly_max
```

```python
# What is the frequency of the time series? 
monthly_max.index.freq
```

```python
# Make a plot of the daily mean NO2 value, in each of the stations. 
no_2.resample("D").mean().plot(style = "-o", figsize = (10, 5))
plt.ylabel("Daily Average $NO_2$")
plt.show()
```
