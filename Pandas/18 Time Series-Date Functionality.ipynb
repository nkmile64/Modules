## Custom Business Days

```python
# In Egypt a Friday-Saturday weekend is observed
weekmask_egypt = "Sun Mon Tue Wed Thu"

# They alos observe International Worker's Day so let's add that for a couple of years
holidays = [
    "2012-05-01",
    datetime.datetime(2013, 5, 1),
    np.datetime64("2014-05-01"),
]

bday_egypt = pd.offsets.CustomBusinessDay(holidays=holidays, weekmask=weekmask_egypt)
dt = datetime.datetime(2013, 4, 30)
dt + 2 * bday_egypt
```

```python
# Map to the weekday names:
dts = pd.date_range(dt, periods=5, freq=bday_egypt)
pd.Series(dts.weekday, dts).map(pd.Series("Mon Tue Wed Thu Fri Sat Sun".split()))
```

```python
from pandas.tseries.holiday import USFederalHolidayCalendar

bday_us = pd.offsets.CustomBusinessDay(calendar=USFederalHolidayCalendar())

# Friday before MLK Day
dt = datetime.datetime(2014, 1, 17)

# Tuesday after MLK Day (Monday is skipped because it's a holiday)
dt + bday_us
```

# Resampling
## Basics: Resampling means frequency conversion

```python
rng = pd.date_range("1/1/2012", periods=100, freq="s")
ts = pd.Series(np.random.randint(0, 500, len(rng)), index=rng)
ts.resample("5min").sum()
```

```python
ts.resample("5min").mean()
```

```python
ts.resample("5min").ohlc()
```

```python
ts.resample("5min").max()
```

```python
# Specify that the right side of the bin interval is closed
ts.resample("5min", closed="right").mean()
```

```python
# Specify that the left side of the bin interval is closed
ts.resample("5min", closed="left").mean()
```

## Upsampling

```python
# From secondly to every 250 ms
# Use only the first two values of the original Series i.e. at 00:00:00 and 00:00:01 seconds
# Then upsample to 250ms
ts[:2].resample("250ms").asfreq()
```

```python
# Forward fill with the last non-null value
ts[:2].resample("250ms").ffill()
```

```python
ts[:2].resample("250ms").ffill(limit=2)
```

## Sparse Resampling

```python
rng = pd.date_range("2014-1-1", periods=100, freq="D") + pd.Timedelta("1s")
ts = pd.Series(range(100), index=rng)
ts
```

```python
# Resample to the full range of the series
ts.resample("3min").sum()
```

```python
# Only resample to those groups where we have points
from functools import partial
from pandas.tseries.frequencies import to_offset


def round(t, freq):
    freq = to_offset(freq)
    td = pd.Timedelta(freq)

    return pd.Timestamp((t.value // td.value) * td.value)


ts.groupby(partial(round, freq="3min")).sum()
```

## Iterating through Groups

```python
small = pd.Series(
    range(6),
    index=pd.to_datetime(
        [
            "2017-01-01T00:00:00",
            "2017-01-01T00:30:00",
            "2017-01-01T00:31:00",
            "2017-01-01T01:00:00",
            "2017-01-01T03:00:00",
            "2017-01-01T03:05:00",
        ]
    ),
)

resampled = small.resample("h")

for name, group in resampled:
    print(f"Group: {name}")
    print("-" * 27)
    print(group, end="\n\n")
```
