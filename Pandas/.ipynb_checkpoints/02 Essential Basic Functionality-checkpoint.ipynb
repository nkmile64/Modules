## Combining Overlapping Data Sets

```python
# Combine two DataFrames where missing values in one of them, are 
# conditionally filled with like-labeled values from the other. 
df1 = pd.DataFrame(
    {
        "A": [1.0, np.nan, 3.0, 5.0, np.nan], 
        "B": [np.nan, 2.0, 3.0, np.nan, 6.0]
    }
)
df2 = pd.DataFrame(
    {
        "A": [5.0, 2.0, 4.0, np.nan, 3.0, 7.0], 
        "B": [np.nan, np.nan, 3.0, 4.0, 6.0, 8.0]
    }
)
```

```python
df1
```

```python
df2
```

```python
df1.combine_first(df2)
```

## General DataFrame combine

```python
def combiner(x, y):
    return np.where(pd.isna(x), y, x)

df1.combine(df2, combiner)
```

## Function Application

### Tablewise Function Application
##### DataFrames (and Series) can be passed into functions.
##### If the function will be used in a chain, use the <b>pipe()</b> method.

```python
def extract_city_name(df):
    """Chicago, IL -> Chicago for city_name column""" 

    df["city_name"] = df["city_and_code"].str.split(",").str.get(0) 
    return df 

def add_country_name(df, country_name = None):
    """Chicago -> Chicago-US for city_name column""" 

    col = "city_name" 
    df["city_and_country"] = df[col] + country_name 
    return df 

df_p = pd.DataFrame({"city_and_code": ["Chicago, IL"]})
df_p
```

```python
add_country_name(extract_city_name(df_p), country_name = "US")
```

```python
# the above is equivalent to (method chaining):
df_p.pipe(extract_city_name).pipe(add_country_name, country_name = "US")
```

```python
# Here, the function we wish to apply takes the data as its second argument
import statsmodels.formula.api as sm 

bb = pd.read_csv("data/baseball.csv", index_col = "id")

(
    bb.query("h > 0")
    .assign(ln_h = lambda df: np.log(df.h))
    .pipe((sm.ols, "data"), "hr ~ ln_h + year + g + C(lg)")
    .fit() 
    .summary()
) 
```

### Filling while reindexing

```python
# a simple DatetimeIndex
rng = pd.date_range("1/3/2000", periods = 8)
rng
```

```python
# a Series of 8 random samples from the Standard Normal Distribution
ts = pd.Series(np.random.randn(8), index = rng)
```

```python
# a subset of the time series 
ts2 = ts.iloc[[0, 3, 6]]
```

```python
ts
```

```python
ts2
```

```python
ts2.reindex(ts.index)
```

```python
# fill missing values forward 
ts2.reindex(ts.index, method = "ffill")
```

```python
# fill missing values backwards
ts2.reindex(ts.index, method = "bfill")
```

```python
ts2.reindex(ts.index, method = "nearest")
```

### Limits on filling while reindexing

```python
# specify the maximum count of consecutive matches 
ts2.reindex(ts.index, method = "ffill", limit = 1)
```

```python
# specify maximum distance between index and indexer values 
ts2.reindex(ts.index, method = "ffill", tolerance = "1 day")
```
