# Hierarchical Indexing (MultiIndex)

# Creating a MultiIndex (hierarchical index) Object

```python
# Create a MultiIndex from tuples
arrays = [
    ["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"],
    ["one", "two", "one", "two", "one", "two", "one", "two"],
]

tuples = list(zip(*arrays))
tuples
```

```python
index = pd.MultiIndex.from_tuples(tuples, names=["first", "second"])
index
```

```python
s = pd.Series(np.random.randn(8), index=index)
s
```

```python
# Create a MultiIndex from the cross product of two iterables
# i.e. include every pairing of the elements
iterables = [
    ["bar", "baz", "foo", "qux"],
    ["one", "two"],
]

pd.MultiIndex.from_product(iterables, names=["first", "second"])
```

```python
# Create a MultiIndex from a DataFrame
df = pd.DataFrame(
    [
        ["bar", "one"],
        ["bar", "two"],
        ["foo", "one"],
        ["foo", "two"],
    ],
    columns=["first", "second"],
)

pd.MultiIndex.from_frame(df)
```

```python
# Or pass a list of arrays directly to a Series (or DataFrame)
arrays = [
    np.array(["bar", "bar", "baz", "baz", "foo", "foo", "qux", "qux"]),
    np.array(["one", "two", "one", "two", "one", "two", "one", "two"]),
]

s = pd.Series(np.random.randn(8), index=arrays)
s
```

```python
df = pd.DataFrame(np.random.randn(8, 4), index=arrays)
df
```

```python
# names holds the names of the levels
# if no names argument was provided, None is returned
df.index.names
```

```python
# The MultiIndex we created can be used on any axis of the DataFrame
# Above it was used as row labels (axis=0)
# Below as column labels
df = pd.DataFrame(np.random.randn(3, 8), index=["A", "B", "C"], columns=index)
df
```

```python
pd.DataFrame(np.random.randn(6, 6), index=index[:6], columns=index[:6])
```

## Data Alignment and Using reindex

```python
# When the objects have different indexes (MultiIndexes)
# we have data alignment.
s  # a Series with hierarchical index
```

```python
# Add two Series element-wise. The second one has elements
# from the start up to (qux, one) - not included.
# So, for example (qux, one) i.e. 0.502 + Nan -> Nan
s + s[:-2]
```

```python
# step 2: include (bar, one), (baz, one), (foo, one) and (qux, one).
#       : so for example (bar, two): -1.905 + NaN -> NaN
s + s[::2]
```

```python
# Conform the Series to a new index
s.reindex(index[:3])
```

```python
# Use a list of tuples for re-indexing
s.reindex([("foo", "two"), ("bar", "one"), ("qux", "one"), ("baz", "one")])
```

## Using Slicers

```python
def mklbl(prefix, n):
    return ["%s%s" % (prefix, i) for i in range(n)]


miindex = pd.MultiIndex.from_product(
    [mklbl("A", 4), mklbl("B", 2), mklbl("C", 4), mklbl("D", 2)],
)
micolumns = pd.MultiIndex.from_tuples(
    [("a", "foo"), ("a", "bar"), ("b", "foo"), ("b", "bah")],
    names=["lvl0", "lvl1"],
)

dfmi = (
    pd.DataFrame(
        np.arange(len(miindex) * len(micolumns)).reshape(len(miindex), len(micolumns)),
        index=miindex,
        columns=micolumns,
    )
    .sort_index()
    .sort_index(axis=1)
)

dfmi
```

```python
# Basic MultiIndex slicing using slices, lists & labels
dfmi.loc[(slice("A1", "A3"), slice(None), ["C1", "C3"]), :]
```

```python
idx = pd.IndexSlice
dfmi.loc[idx[:, :, ["C1", "C3"]], idx[:, "foo"]]
```

```python
dfmi.loc["A1", (slice(None), "foo")]
```

```python
dfmi.loc[idx[:, :, ["C1", "C3"]], idx[:, "foo"]]
```

```python
# Boolean indexer
mask = dfmi[("a", "foo")] > 200
dfmi.loc[idx[mask, :, ["C1", "C3"]], idx[:, "foo"]]
```

```python
# Using the "axis" argument
dfmi.loc(axis=0)[:, :, ["C1", "C3"]]
```

```python
# Set the values
df2 = dfmi.copy()
df2.loc(axis=0)[:, :, ["C1", "C3"]] = -10
df2
```

```python
df2 = dfmi.copy()
df2.loc[idx[:, :, ["C1", "C3"]], :] = df2 * 1000
df2
```

## Advanced Reindexing and Alignment

```python
# Broadcast values across a level.
midx = pd.MultiIndex(
    levels=[["zero", "one"], ["x", "y"]],
    codes=[[1, 1, 0, 0], [1, 0, 1, 0]],
)

df = pd.DataFrame(np.random.randn(4, 2), index=midx)
df
```

```python
df2 = df.groupby(level=0).mean()
df2
```

```python
df2.reindex(df.index, level=0)
```

```python
# aligning
df_aligned, df2_aligned = df.align(df2, level=0)
df_aligned
```

```python
df2_aligned
```

## CategoricalIndex

Allows efficient indexing of an index with a large number of <b><u>duplicated</u></b> elements.

```python
from pandas.api.types import CategoricalDtype

df = pd.DataFrame({"A": np.arange(6), "B": list("aabbca")})

df["B"] = df["B"].astype(CategoricalDtype(list("cab")))

df
```

```python
df.dtypes
```

```python
df["B"].cat.categories
```

```python
# Set as index the existing column "B"
df2 = df.set_index("B")

df2.index
```

```python
df2.loc["a"]
```

```python
df2.loc["a"].index
```

```python
# Sort the index by the order of the categories as we specified it!
df2.sort_index()
```

```python
df2.groupby(level=0, observed=True).sum().index
```
