# Basics

```python
dates = pd.date_range("1/1/2000", periods=8)
df = pd.DataFrame(
    np.random.randn(8, 4),
    index=dates,
    columns=["A", "B", "C", "D"],
)
```

```python
df
```

```python
# Indexing with [] on a DataFrame returns the column values as a Series
s = df["A"]

# Indexing with [] on a Series returns a scalar
s[dates[5]]
```

```python
# Exchange values between columns A and B
df[["B", "A"]] = df[["A", "B"]]
df
```

```python
# WARNING: Using .loc always aligns the AXES. So, trying to make a
#          transform with .loc will fail because column alignment
#          happens before value assignment!
df[["A", "B"]]
```

```python
df.loc[:, ["B", "A"]] = df[["A", "B"]]
df[["A", "B"]]
```

```python
# To swap column values with .loc, use raw values
df.loc[:, ["B", "A"]] = df[["A", "B"]].to_numpy()
df[["A", "B"]]
```

```python
# NOTE: When using .iloc pandas DOES NOT align axes so
#       the following will swap the columns
df[["A", "B"]]
```

```python
df.iloc[:, [1, 0]] = df[["A", "B"]]
df[["A", "B"]]
```

# Selecting Random Samples

```python
s = pd.Series([0, 1, 2, 3, 4, 5])

# If no arguments passed, returns 1 row
s.sample()
```

```python
# Can specify either a number of rows
s.sample(n=3)
```

```python
# or a fraction of the rows
s.sample(frac=0.5)
```

```python
# Sample with replacement
s = pd.Series([0, 1, 2, 3, 4, 5])

# Without replacement - the default
s.sample(n=6, replace=False)
```

```python
# with replacement
s.sample(n=6, replace=True)
```

```python
# If you want rows to have different probabilities of being selected,
# pass the sampling function sampling weights.
s = pd.Series([0, 1, 2, 3, 4, 5])
example_weight = [0, 0, 0.2, 0.2, 0.2, 0.4]
s.sample(n=3, replace=False, weights=example_weight)
```

```python
# weights will be re-normalized automatically
example_weight2 = [0.5, 0, 0, 0, 0, 0]
s.sample(n=1, replace=False, weights=example_weight2)
```

```python
# When using a DataFrame, pass as a string the name of the column with the weights
df2 = pd.DataFrame({"col1": [9, 8, 7, 6], "weight_column": [0.5, 0.4, 0.1, 0]})

df2.sample(n=3, weights="weight_column")
```

```python
# Can also sample columns instead of rows
df3 = pd.DataFrame({"col1": [1, 2, 3], "col2": [2, 3, 4]})

df3.sample(n=1, axis=1)
```

```python
df4 = pd.DataFrame({"col1": [1, 2, 3], "col2": [2, 3, 4]})

# With a given seed, the sample will always draw the same rows
df4.sample(n=2, random_state=2)
```

# Setting with Enlargement Conditionally Using numpy()

```python
df = pd.DataFrame(
    {
        "col1": list("ABBC"),
        "col2": list("ZZXY"),
    }
)

# Set a new column color to "green" when the second column has "Z"
df["color"] = np.where(df["col2"] == "Z", "green", "red")
df
```

```python
# For multiple conditions, use np.select()
conditions = [
    (df["col2"] == "Z") & (df["col1"] == "A"),
    (df["col2"] == "Z") & (df["col1"] == "B"),
    (df["col1"] == "B"),
]

choices = ["yellow", "blue", "purple"]
df["color"] = np.select(conditions, choices, default="black")
df
```

## Boolean Operators

```python
# not/~ negates a boolean expression
df = pd.DataFrame(np.random.randn(n, 3), columns=list("abc"))
df["bools"] = np.random.randn(len(df)) > 0.5
df.query("~bools")
```

```python
df.query("not bools")
```

```python
df.query("not bools") == df[~df["bools"]]
```

```python
# Short query syntax
shorter = df.query("a < b < c and (not bools) or bools > 2")
shorter
```

```python
# In pure Python
longer = df[(df["a"] < df["b"]) & (df["b"] < df["c"]) & (~df["bools"]) | (df["bools"] > 2)]
longer
```

```python
shorter == longer
```

# Returning a View versus a Copy

```python
dfmi = pd.DataFrame(
    [
        list("abcd"),
        list("efgh"),
        list("ijkl"),
        list("mnop"),
    ],
    columns=pd.MultiIndex.from_product(
        [
            ["one", "two"],
            ["first", "second"],
        ]
    ),
)
dfmi
```

```python
# First access method: chained []
dfmi["one"]["second"]
```

```python
# Second access method: labeled-based location (preferred method)
dfmi.loc[:, ("one", "second")]
```
