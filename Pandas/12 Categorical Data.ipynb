# Object Creation
## Series Creation
- Specifying dtype="category"
- Converting existing Series/DataFrame column, to a `category` dtype
- Using special functions such as `cut()`, which groups data into discrete bins
- By passing a Categorical object to a Series

```python
import pandas as pd
import numpy as np

s = pd.Series(["a", "b", "c", "a"], dtype="category")
s
```

```python
df = pd.DataFrame({"A": ["a", "b", "c", "a"]})
df["B"] = df["A"].astype("category")
df
```

```python
df = pd.DataFrame({"value": np.random.randint(0, 100, 20)})
labels = ["{0} - {1}".format(i, i + 9) for i in range(0, 100, 10)]
labels
```

```python
df["group"] = pd.cut(df.value, range(0, 105, 10), right=False, labels=labels)
df.head(10)
```

```python
raw_cat = pd.Categorical(
    ["a", "b", "c", "a"],
    categories=["b", "c", "d"],
    ordered=False,
)

s = pd.Series(raw_cat)
s
```

```python
df = pd.DataFrame({"A": ["a", "b", "c", "a"]})
df["B"] = raw_cat
df
```

```python
df.dtypes
```

# CategoricalDtype

```python
from pandas.api.types import CategoricalDtype

# The categories are inferred from the data, and assumed to be unordered
CategoricalDtype(["a", "b", "c"])
```

```python
# Categories again are inferred, but now they are ordered
CategoricalDtype(["a", "b", "c"], ordered=True)
```

```python
CategoricalDtype()
```

## Equality Semantics

```python
# Two unordered categoricals with the same categories, are equal regardless of the order
c1 = CategoricalDtype(["a", "b", "c"], ordered=False)
c2 = CategoricalDtype(["b", "c", "a"], ordered=False)
c1 == c2
```

```python
# One unordered and one ordered categorical with the same categories, are compared unequal
c1 = CategoricalDtype(["a", "b", "c"], ordered=False)
c2 = CategoricalDtype(["a", "b", "c"], ordered=True)
c1 == c2
```

```python
# Also note:
c1 == "category"
```

## Setting

```python
# To set a value in a categorical Series, the value must be included in the categories
idx = pd.Index(["h", "i", "j", "k", "l", "m", "n"])
cats = pd.Categorical(["a", "a", "a", "a", "a", "a", "a"], categories=["a", "b"])
values = [1, 1, 1, 1, 1, 1, 1]
df = pd.DataFrame({"cats": cats, "values": values}, index=idx)
df
```

```python
df.iloc[2:4, :] = [["b", 2], ["b", 2]]
df
```

```python
# Value "c" is not included in "categories", so it raises:
try:
    df.iloc[2:4, :] = [["c", 3], ["c", 3]]
except TypeError as e:
    print("TypeError", str(e))
```

## Merging / Concatenation

```python
from pandas.api.types import union_categoricals

# same categories
s1 = pd.Series(["a", "b"], dtype="category")
s2 = pd.Series(["a", "b", "a"], dtype="category")
pd.concat([s1, s2])
```

```python
# different categories: the results now are "object" and not "category"
s3 = pd.Series(["b", "c"], dtype="category")
pd.concat([s1, s3])
```

```python
# output type is inferred based on categories values
int_cats = pd.Series([1, 2], dtype="category")
float_cats = pd.Series([3.0, 4.0], dtype="category")
pd.concat([int_cats, float_cats])
```

```python
union_categoricals([s1.array, s3.array])
```

## Unioning

```python
# To combine categoricals that don't necessarily have the same categories:
from pandas.api.types import union_categoricals

a = pd.Categorical(["b", "c"])
b = pd.Categorical(["a", "b"])

# Return the union of the categories as a list:
union_categoricals([a, b])
```

```python
# For lexical sorting of the categories:
union_categoricals([a, b], sort_categories=True)
```

```python
a = pd.Categorical(["a", "b"], ordered=True)
b = pd.Categorical(["a", "b", "a"], ordered=True)

union_categoricals([a, b])
```

```python
# For not identical but ordered categories, an error is raised:
try:
    a = pd.Categorical(["a", "b"], ordered=True)
    b = pd.Categorical(["a", "b", "c"], ordered=True)
    union_categoricals([a, b])
except TypeError as e:
    print("TypeError: ", str(e))
```

```python
# To avoid the above error, use ignore_ordered=True:
a = pd.Categorical(["a", "b"], ordered=True)
b = pd.Categorical(["a", "b", "c"], ordered=True)
union_categoricals([a, b], ignore_order=True)
```

```python
a = pd.Series(["b", "c"], dtype="category")
b = pd.Series(["a", "b"], dtype="category")

union_categoricals([a, b])
```

# Getting Data In/Out

```python
# If we write and later read back data from a .csv file, we must convert the
# relevant columns back to "category"
import io

s = pd.Series(pd.Categorical(["a", "b", "b", "a", "a", "d"]))

# Rename the categories
s = s.cat.rename_categories(["very good", "good", "bad"])

# Re-order the categories and add missing categories
s = s.cat.set_categories(["very bad", "bad", "medium", "good", "very good"])

df = pd.DataFrame({"cats": s, "vals": [1, 2, 3, 4, 5, 6]})

csv = io.StringIO()
df.to_csv(csv)  # convert to CSV but in memory
df2 = pd.read_csv(io.StringIO(csv.getvalue()))
df2.dtypes
```

```python
# The "cats" column, when read from the CSV, is no longer of "category" dtype
df2["cats"]
```

```python
# Redo the category
df2["cats"] = df2["cats"].astype("category")

# Re-order and set missing categories
df2["cats"] = df2["cats"].cat.set_categories(["very bad", "bad", "medium", "good", "very good"])

# Now check the dtype
df2.dtypes
```

# Categorical is not a `numpy` array
