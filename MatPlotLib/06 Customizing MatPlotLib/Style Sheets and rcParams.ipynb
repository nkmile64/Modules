# Runtime rc settings

Dynamically change the default rc (runtime configuration) settings through `rcParams`; a dictionary-like variable that is global to the matplotlib package.

```python
from cycler import cycler
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

plt.rcParams["lines.linewidth"] = 2
plt.rcParams["lines.linestyle"] = "--"

data = np.random.randn(50)
plt.plot(data)

plt.show()
```

```python
# Change the usual plot color
mpl.rcParams["axes.prop_cycle"] = cycler(color=["r", "g", "b", "y"])
plt.plot(data)  # first color is red

plt.show()
```

`rc` can be used to modify multiple settings in a single group at once.<br/>
The group could be "lines", "figure", "axes" etc.

```python
mpl.rc("lines", linewidth=4, linestyle="-.")
plt.plot(data)

plt.show()
```

## Temporary rc Settings

Runtime configuration settings can be changed temporarily with the `rc_context` manager:

```python
with plt.rc_context({"lines.linewidth": 2, "lines.linestyle": ":"}):
    plt.plot(data)
```

It can also be used as a **decorator** within a function:

```python
@mpl.rc_context({"lines.linewidth": 4, "lines.linestyle": "-."})
def plotting_function():
    plt.plot(data)


plotting_function()
```

```python
# Restore rc from matplotlib's internal style sheet
mpl.rcdefaults()
```

```python
plt.style.use("ggplot")
plt.plot(data)
```
