```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)
s = np.sin(2 * np.pi * t)

fig, ax = plt.subplots()

ax.plot(t, s, color="black")
ax.axhline(0, color="black")

# Fill the area between two horizontal lines
ax.fill_between(t, 1, where=s > 0, fc="g", alpha=0.5)
ax.fill_between(t, -1, where=s < 0, fc="r", alpha=0.5)

plt.show()
```
