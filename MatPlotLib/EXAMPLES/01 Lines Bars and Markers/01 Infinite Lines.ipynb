#### Infinite Lines

```python
import matplotlib.pyplot as plt
import numpy as np

# Return 100 evenly spaced samples in the interval [-10, 10]
t = np.linspace(-10, 10, 100)

# The Sigmoid function
sig = 1 / (1 + np.exp(-t))

fig, ax = plt.subplots(figsize=(10, 6))

# Infinite horizontal line: spans the whole of the Axes
ax.axhline(y=0, color="black", linestyle="--")  # mark the lower limit
ax.axhline(y=0.5, color="brown", linestyle=":")  # mark the center
ax.axhline(y=1.0, color="black", linestyle="--")  # mark the upper limit

ax.axvline(color="grey")

# An arbitrary straight line that has to pass through
# point (0, 0.5) and has a slope of 0.25
ax.axline((0, 0.5), slope=0.25, color="red", linestyle=(0, (5, 5)))

ax.plot(t, sig, linewidth=2, label=r"$\sigma(t) = \frac{1}{1 + e^{-t}}$")
ax.set(xlim=(-10, 10), xlabel="t")
ax.legend(fontsize=14)

plt.show()
```

#### Drawing Diagonal Grid Lines

```python
fig, ax = plt.subplots()

for pos in np.linspace(-2, 1, 10):
    ax.axline((pos, 0), slope=0.5, color="k", transform=ax.transAxes)


ax.set(xlim=(0, 1), ylim=(0, 1))

plt.show()
```
