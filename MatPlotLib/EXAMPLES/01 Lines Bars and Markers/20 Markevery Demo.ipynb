**Line2D.markevery** : draws markers at a subset of data points.

> - **N** draws every Nth marker
> - **(start, N)** draws every Nth marker starting at *start*
> - **[]** draws markers at specified indices
> - **slice** draws markers at the sliced indices
> - **float** distance between markers as a fraction of the Axes diagonal; creates a uniform distribution



```python
import matplotlib.pyplot as plt
import numpy as np

# define a list of markevery cases to plot
cases = [
    None,
    8,
    (30, 8),
    [16, 24, 32],
    [0, -1],
    slice(100, 200, 3),
    0.1,
    0.4,
    (0.2, 0.4),
]

# data points
delta = 0.11
x = np.linspace(0, 10 - 2 * delta, 200) + delta
y = np.sin(x) + 1.0 + delta
```

# markevery with Linear Scales

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 6), layout="constrained")
for ax, markevery in zip(axs.flat, cases):
    ax.plot(x, y, marker="o", ls="-", ms=4, markevery=markevery)
    ax.set_title(f"markevery={markevery}")
```

# markevery with Log Scales

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 6), layout="constrained")

for ax, markevery in zip(axs.flat, cases):
    ax.plot(x, y, marker="o", ls="-", ms=4, markevery=markevery)
    ax.set_title(f"markevery={markevery}")
    ax.set_xscale("log")
    ax.set_yscale("log")
```

# markevery on Zoomed Plots

```python
fig, axs = plt.subplots(3, 3, figsize=(10, 6), layout="constrained")

for ax, markevery in zip(axs.flat, cases):
    ax.plot(x, y, marker="o", ls="-", ms=4, markevery=markevery)
    ax.set_title(f"markevery={markevery}")
    ax.set_xlim((6, 6.7))
    ax.set_ylim((1.1, 1.7))
```

# markevery on Polar Plots

```python
r = np.linspace(0, 3.0, 200)
theta = 2 * np.pi * r

fig, axs = plt.subplots(
    3,
    3,
    figsize=(10, 6),
    layout="constrained",
    subplot_kw={"projection": "polar"},
)

for ax, markevery in zip(axs.flat, cases):
    ax.plot(theta, r, marker="o", ls="-", ms=4, markevery=markevery)
    ax.set_title(f"markevery={markevery}")

plt.show()
```
