```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 500)
y = np.sin(x)

plt.rc("lines", linewidth=2.5)
fig, ax = plt.subplots(figsize=(10, 6))

# Using set_dashes() and set_capstyle() to modify dashing on an existing line.
(line1,) = ax.plot(x, y, label="Using set_dashes() and set_capstyle()")
# a sequence of 2-point and 10-point dashes, separated by 2-point spaces.
line1.set_dashes([2, 2, 10, 2])
line1.set_dash_capstyle("round")

# Using plot(..., dashes=...) to set the dashing when creating a line.
(line2,) = ax.plot(x, y - 0.2, dashes=[6, 2], label="Using the dashes argument")

# Using plot(..., dashes=..., gapcolor=...) to set dashing and alternating
# color when creating a line.
(line3,) = ax.plot(x, y - 0.4, dashes=[4, 4], gapcolor="tab:pink", label="Using the dashes and gapcolor arguments")

ax.legend(handlelength=4)
plt.show()
```
