```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, layout="constrained")

ax = axs[0][0]
ax.plot(np.arange(0, 1e6, 1000))
ax.set_title("Title [0 0]")
ax.set_ylabel("YLabel [0 0]")

ax = axs[0][1]
ax.plot(np.arange(1.0, 0.0, -0.1) * 2000.0, np.arange(1.0, 0.0, -0.1))
ax.set_title("Title [0 1]")
ax.xaxis.tick_top()
ax.tick_params(axis="x", rotation=55)

for i in range(2):
    ax = axs[1][i]
    ax.plot(np.arange(1.0, 0.0, -0.1) * 2000.0, np.arange(1.0, 0.0, -0.1))
    ax.set_ylabel(f"YLabel [1 {i}]")
    ax.set_xlabel(f"XLabel [1 {i}]")
    if i == 0:
        ax.tick_params(axis="x", rotation=55)

fig.align_labels()
fig.align_titles()

plt.show()
```
