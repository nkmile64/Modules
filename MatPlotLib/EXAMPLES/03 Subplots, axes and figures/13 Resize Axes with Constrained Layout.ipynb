With no **constrained layout**, the labels overlap the Axes.

```python
import matplotlib.pyplot as plt


def example_plot(ax: plt.axes._axes.Axes):
    ax.plot([1, 2])
    ax.set_xlabel("x-label", fontsize=12)
    ax.set_ylabel("y-label", fontsize=12)
    ax.set_title("Title", fontsize=14)


fig, axs = plt.subplots(nrows=2, ncols=2, layout=None)

for ax in axs.flat:
    example_plot(ax)
```

Add **layout="constrained"** to automatically adjust  and avoid overlap.

```python
fig, axs = plt.subplots(2, 2, layout="constrained")

for ax in axs.flat:
    example_plot(ax)
```

More complicated example using **nested gridspecs**.

```python
import matplotlib.gridspec as gridspec

fig = plt.figure(layout="constrained")

gs0 = gridspec.GridSpec(1, 2, figure=fig)

# First column of gs0
gs1 = gridspec.GridSpecFromSubplotSpec(3, 1, subplot_spec=gs0[0])
for n in range(3):
    ax = fig.add_subplot(gs1[n])
    example_plot(ax)

# Second column of gs0
gs2 = gridspec.GridSpecFromSubplotSpec(2, 1, subplot_spec=gs0[1])
for n in range(2):
    ax = fig.add_subplot(gs2[n])
    example_plot(ax)

plt.show()
```
