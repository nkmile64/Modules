```python
import itertools
import warnings
import matplotlib.pyplot as plt

fontsizes = itertools.cycle([8, 16, 24, 32])


def example_plot(ax: plt.axes._axes.Axes):
    ax.plot([1, 2])
    ax.set_xlabel("X label", fontsize=next(fontsizes))
    ax.set_ylabel("Y label", fontsize=next(fontsizes))
    ax.set_title("Title", fontsize=next(fontsizes))
```

```python
fig, ax = plt.subplots()
example_plot(ax)
fig.tight_layout()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2)

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)

fig.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)

example_plot(ax1)
example_plot(ax2)

fig.tight_layout()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

example_plot(ax1)
example_plot(ax2)

fig.tight_layout()
```

```python
fig, axs = plt.subplots(nrows=3, ncols=3)

for ax in axs.flat:
    example_plot(ax)

fig.tight_layout()
```

```python
plt.figure()

ax1 = plt.subplot(221)
ax2 = plt.subplot(223)
ax3 = plt.subplot(122)

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)

plt.tight_layout()
```

```python
plt.figure()

# subplot2grid(tuple:num of rows/columns of the grid into which to place subplot,
#              tuple:row number, column number of the location)
ax1 = plt.subplot2grid((3, 3), (0, 0))
ax2 = plt.subplot2grid((3, 3), (0, 1), colspan=2)
ax3 = plt.subplot2grid((3, 3), (1, 0), colspan=2, rowspan=2)
ax4 = plt.subplot2grid((3, 3), (1, 2), rowspan=2)

example_plot(ax1)
example_plot(ax2)
example_plot(ax3)
example_plot(ax4)

plt.tight_layout()
```
