```python
# Define the imports
import matplotlib.pyplot as plt
import numpy as np

# Create a figure with subplots
fig, ax = plt.subplots(figsize=(8, 5))

recipe = [
    "375 g flour",
    "75 g sugar",
    "250 g butter",
    "300 g berries",
]

# Isolate the numbers of grams - the data
data = [int(x.split()[0]) for x in recipe]
# Isolate the names of the ingredients - labels
ingredients = [x.split()[-1] for x in recipe]

# Get the PieContainer object
pie = ax.pie(data)

# Label each ingredient's wedge with percentages and its weight in grams
ax.pie_label(
    pie,    # container with all the wedges
    "{frac:.1%}\n({absval:d}g)",    # the labels for each wedge
    textprops=dict(
        color="w",
        size=10,
        weight="bold",
    ),
)
ax.legend(
    pie.wedges, # use the pie's wedges, as the handles for the legend
    ingredients, # use the ingredients as the labels for the legend
    title="Ingredients",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
)
ax.set_title("Matplotlib bakery: A pie")

plt.show()
```
