<h4 style="color:#a59f4b">Label Slices</h4>

```python
import matplotlib.pyplot as plt

labels = ["Frogs", "Hogs", "Dogs", "Logs"]
sizes = [15, 30, 45, 10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels)

plt.show()
```

<h4 style="color:#a59f4b">Auto-Label Slices with their numeric values</h4>

```python
fig, ax = plt.subplots()

ax.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.show()
```

<h4 style="color:#a59f4b;font-weight:600">Color Slices</h4>

```python
sizes = [15, 30, 45, 10]
labels = ["Frogs", "Hogs", "Dogs", "Logs"]
colors = ["olivedrab", "rosybrown", "gray", "saddlebrown"]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors)

plt.show()
```

<h4 style="color:#a59f4b;font-weight:600">Hatch Slices</h4>

```python
sizes = [15, 30, 45, 10]
labels = ["Frogs", "Hogs", "Dogs", "Logs"]
colors = ["olivedrab", "rosybrown", "gray", "saddlebrown"]
hatch = ["**O", "oO", "O.O", ".||."]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, hatch=hatch)

plt.show()
```

<h4 style="color:#a59f4b;font-weight:600">Swap Label and Autopct Text Positions</h4>

```python
fig, ax = plt.subplots()

ax.pie(sizes, labels=labels, autopct="%1.1f%%", pctdistance=1.25, labeldistance=0.6)
plt.show()
```

<h4 style="color:#a59f4b;font-weight:600">Explode, Shade, and Rotate Slices</h4>

```python
# Only "explode" the 2nd slice (i.e. Hogs)
explode = (0, 0.1, 0, 0)

fig, ax = plt.subplots(figsize=(6.5, 4.5), layout="constrained")
ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90,
)
plt.show()
```

<h4 style="color:#a59f4b;font-weight:600">Controlling the Size</h4>

```python
fig, ax = plt.subplots()

ax.pie(
    sizes,
    labels=labels,
    autopct="%1.0f%%",
    textprops={"size": "smaller"},
    radius=0.4,
)
plt.show()
```

<h4 style="color:#a59f4b;font-weight:600">Modifying the Shadow</h4>

```python
fig, ax = plt.subplots()
ax.pie(
    sizes,
    explode=explode,
    labels=labels,
    autopct="%1.1f%%",
    shadow={"ox": -0.04, "edgecolor": "none", "shade": 0.9},
    startangle=90,
)
plt.show()
```
