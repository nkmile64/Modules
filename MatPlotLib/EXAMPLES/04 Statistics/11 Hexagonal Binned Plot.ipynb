`hexbin` is a 2D **histogram** plot in which:
- the bins are hexagons and
- the color represents the number of data points within each bin

```python
import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)

n = 100_000
x = np.random.standard_normal(n)
y = 2.0 + 3.0 * x + 4.0 * np.random.standard_normal(n)  # type: ignore
xlim = x.min(), x.max()
ylim = y.min(), y.max()

fig, (ax0, ax1) = plt.subplots(ncols=2, sharey=True, figsize=(11, 5), layout="constrained")

hb = ax0.hexbin(x, y, gridsize=50, cmap="inferno")
ax0.set(xlim=xlim, ylim=ylim)
ax0.set_title("Hexagon binning")
cb = fig.colorbar(hb, ax=ax0, label="counts")

hb = ax1.hexbin(x, y, gridsize=50, cmap="inferno", bins="log")
ax1.set(xlim=xlim, ylim=ylim)
ax1.set_title("With a log color scale")
cb = fig.colorbar(hb, ax=ax1, label="counts")

plt.show()
```
