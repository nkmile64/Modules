```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)

n_bins = 10
x = np.random.randn(1000, 3)

fig, ((ax0, ax1), (ax2, ax3)) = plt.subplots(2, 2, figsize=(9, 6))

colors = ["red", "tan", "lime"]
ax0.hist(x, n_bins, density=True, histtype="bar", color=colors, label=colors)
ax0.legend(prop={"size": 10})
ax0.set_title("bars with legend")

ax1.hist(x, n_bins, density=True, histtype="bar", stacked=True)
ax1.set_title("stacked bar")

ax2.hist(x, n_bins, histtype="step", stacked=True, fill=False)
ax2.set_title("stacked step (unfilled)")

# Make a multiple-histogram of data-sets with different length.
x_multi = [np.random.randn(n) for n in [10000, 5000, 2000]]
ax3.hist(x_multi, n_bins, histtype="bar")
ax3.set_title("different sample sizes")

fig.tight_layout()
plt.show()
```

### Setting Properties for each Dataset
#### **edge color**

```python
fig, ax = plt.subplots(figsize=(8, 5), layout="constrained")

edgecolors = ["green", "red", "blue"]
ax.hist(x, n_bins, fill=False, histtype="step", stacked=True, edgecolor=edgecolors, label=edgecolors)
ax.legend()
ax.set_title("Stacked Steps with Edge Colors")

plt.show()
```

#### **facecolor**

```python
fig, ax = plt.subplots(figsize=(8, 5), layout="constrained")

facecolors = ["green", "red", "blue"]
ax.hist(x, n_bins, histtype="barstacked", facecolor=facecolors, label=facecolors)
ax.legend()
ax.set_title("Bars with Different Facecolors")

plt.show()
```

#### **hatch**

```python
fig, ax = plt.subplots(figsize=(8, 5), layout="constrained")

hatches = [".", "o", "x"]
ax.hist(x, n_bins, histtype="barstacked", hatch=hatches, label=hatches)
ax.legend()
ax.set_title("Hatches on Stacked Bars")

plt.show()
```

#### **linewidth**

```python
fig, ax = plt.subplots(figsize=(8, 5), layout="constrained")

linewidths = [1, 2, 3]
edgecolors = ["green", "red", "blue"]

ax.hist(x, n_bins, fill=False, histtype="bar", linewidth=linewidths, edgecolor=edgecolors, label=edgecolors)
ax.legend()
ax.set_title("Bars with Linewidths")

plt.show()
```

#### **linestyle**

```python
fig, ax = plt.subplots(figsize=(8, 5), layout="constrained")

linestyles = ["-", ":", "--"]

ax.hist(x, n_bins, fill=False, histtype="bar", linestyle=linestyles, edgecolor=edgecolors, label=edgecolors)
ax.legend()
ax.set_title("Bars with Linestyles")

plt.show()
```
