# Creating Figures

```python
# Create a simple 2X2 grid of Axes
fig, axs = plt.subplots(2, 2, figsize=(4, 3), layout="constrained")
```

```python
# For more complex grids:
fig, axs = plt.subplot_mosaic(
    [["A", "right"], ["B", "right"]],
    figsize=(4, 3),
    layout="constrained",
)

for ax_name, ax in axs.items():
    ax.text(0.5, 0.5, ax_name, ha="center", va="center")
```

```python
# To create virtual figures nested inside a parent Figure, use add_subfigure/subfigures
fig = plt.figure(facecolor="lightskyblue", layout="constrained")
figL, figR = fig.subfigures(1, 2)

fig.suptitle("Figure")
figL.suptitle("Left subfigure")
figR.suptitle("Right subfigure")

figL.set_facecolor("thistle")
figR.set_facecolor("paleturquoise")

axL = figL.subplots(2, 1, sharex=True)
axL[1].set_xlabel("x [m]")

axR = figR.subplots(1, 2, sharey=True)
axR[0].set_title("Axes 1")

plt.show()
```

# Matplotlib Application Interfaces (APIs)
## The Explicit "Axes" Interface
1. Create an instance of a <b style="color:yellow;">Figure</b> class
2. Create one (or more) <b style="color:yellow;">Axes</b> objects using subplots()
3. Call drawing methods on the Axes object

```python
fig = plt.figure()
ax = fig.subplots()
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])

plt.show()
```

## The Implicit "pyplot" Interface
Creation of the <b style="color:yellow;">Figure</b> and the <b style="color:yellow;">Axes</b> is done for the user.

```python
plt.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
plt.show()
```

```python
# Using the Implicit method we can get access to the
# current Figure and Axes with: gcf and gca
plt.subplot(1, 2, 1)
ax = plt.gca()
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])

plt.subplot(1, 2, 2)
ax = plt.gca()
ax.plot([3, 2, 1], [0, 0.5, 0.2])
```

```python
# Using the Explicit Interface:
fig, axs = plt.subplots(1, 2)
axs[0].plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
axs[1].plot([3, 2, 1], [0, 0.5, 0.2])
```

# Interactive Figures

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
(ln,) = ax.plot(range(5))
ln.set_color("orange")
plt.ioff()
plt.ion()
plt.isinteractive()

plt.show()
```

```python
# RUN THE .py PROGRAM FOR FULL INTERACTIVITY
from matplotlib.backend_bases import MouseEvent
import matplotlib.pyplot as plt
import numpy as np


# The callback function
def onclick(event: MouseEvent):
    print(f"{'double' if event.dblclick else 'single'} click: button={event.button}", end="\t")
    print(f"x={event.x}", end="\t")
    print(f"y={event.y}", end="\t")
    print(f"xdata={event.xdata}", end="\t")
    print(f"ydata={event.ydata}")


def draw():
    fig, ax = plt.subplots()
    ax.plot(np.random.rand(10))

    # Connect the callback function to the event manager.
    cid = fig.canvas.mpl_connect("button_press_event", onclick)

    # Wait for all figures to be closed, before returning
    plt.show()


draw()
```
