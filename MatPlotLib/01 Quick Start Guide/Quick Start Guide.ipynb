```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# A Simple Example
- <b>Figure</b> : a window containing one or more Axes
- <b>Axes</b> : an area where points can be specified in terms of (x,y) coordinates

```python
fig, ax = plt.subplots()  # create a Figure containing a single Axes
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # plt some data on the Axes
plt.show()  # show the figure
```

# Parts of a Figure

<dd>
    <dt><b>Figure</b>:</dt>
    <dd>Keeps track of all the child Axes, titles, legends etc</dd>
    <dt><b>Axes</b>:</dt>
    <dd>An Artist attached to a Figure, containing a region for plotting data.<br/>
    It usually contains 2 Axis objects each providing ticks and tick labels.<br>
    An Axes has a title, an xlabel and a ylabel.</dd>
    <dt><b>Axis</b>:</dt>
    <dd>Sets scale, limits, ticks(Locator) and ticklabels(Formatter).</dd>
    <dt><b>Artist</b>:</dt>
    <dd>Everything visible on the Figure: Axes, Axis, Text, Line2D etc.</dd>
    
</dd>

# Types of Inputs to Plotting Functions

```python
np.random.seed(19680801)  # seed the random number generator
data = {
    "a": np.arange(50),
    "c": np.random.randint(0, 50, 50),
    "d": np.random.randn(50),
}
data["b"] = data["a"] + 10 * np.random.randn(50)
data["d"] = np.abs(data["d"]) * 100

pd.DataFrame(data).head(5)
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout="constrained")
ax.scatter("a", "b", c="c", s="d", data=data)
ax.set_xlabel("entry a")
ax.set_ylabel("entry b")
```

# Coding Styles
## The Explicit and the Implicit Interfaces

### The Object-Oriented (OO) Style

```python
x = np.linspace(0, 2, 100)  # gets 100 evenly spaced numbers from 0 to 2

fig, ax = plt.subplots(figsize=(5, 2.7), layout="constrained")
ax.plot(x, x, label="linear")  # plot some data on the Axes
ax.plot(x, x**2, label="quadratic")  # plot more data on the Axes
ax.plot(x, x**3, label="cubic")  # ...and some more

ax.set_xlabel("x label")
ax.set_ylabel("y label")
ax.set_title("Simple Plot")
ax.legend()

plt.show()
```

### The pyplot-style

```python
x = np.linspace(0, 2, 100)

plt.figure(figsize=(5, 2.7), layout="constrained")
plt.plot(x, x, label="linear")
plt.plot(x, x**2, label="quadratic")
plt.plot(x, x**3, label="cubic")
plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Simple Plot")
plt.legend()

plt.show()
```

## Making a helper functions

```python
def my_plotter(ax, data1, data2, param_dict):
    """A helper function to make a graph"""

    out = ax.plot(data1, data2, **param_dict)
    return out


# Make 4 random datasets
data1, data2, data3, data4 = np.random.randn(4, 100)

# Create a Figure with a 1X2 grid of Axes
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(5, 2.7))

# Use the helper function to populate two subplots
my_plotter(ax1, data1, data2, {"marker": "x"})
my_plotter(ax2, data3, data4, {"marker": "o"})

plt.show()
```

# Styling Artists

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
x = np.arange(len(data1))

ax.plot(
    x,
    np.cumsum(data1),
    color="blue",
    linewidth=3,
    linestyle="--",
)

# NOTE: plot() returns one or more Line2D objects so
#       we use that to set the linestyle separately.
(l,) = ax.plot(
    x,
    np.cumsum(data2),
    color="orange",
    linewidth=2,
)
l.set_linestyle(":")

plt.show()
```

## Colors

```python
# In a scatter diagram, the edge of the marker can be
# a different color than the interior.
fig, ax = plt.subplots(figsize=(5, 2.7))
ax.scatter(data1, data2, s=50, facecolor="C0", edgecolor="k")

plt.show()
```

## Axes Labels and Test

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10_000)
fig, ax = plt.subplots(figsize=(5, 2.7), layout="constrained")

# The histogram of the data
ax.hist(x, 50, density=True, facecolor="C0", alpha=0.75)
ax.set_xlabel("Length [cm]")
ax.set_ylabel("Probability")
ax.set_title("Aardvark lengths\n (not really)", fontsize=16, color="darkblue")
ax.text(65, 0.025, r"$\mu=115,\ \sigma=15$", fontsize=13, color="red")
ax.axis([55, 175, 0, 0.03])  # min/max values for x-axis and y-axis
ax.grid(True)
plt.show()
```

## Using Mathematical Expressions in Text

## Annotations

```python
fig, ax = plt.subplots(figsize=(5, 2.7))
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
(line,) = ax.plot(t, s, lw=2)

ax.annotate(
    "local max",
    xy=(2, 1),
    xytext=(3, 1.5),
    arrowprops=dict(facecolor="black", shrink=0.05),
)
ax.set_ylim(-2, 2)
```

# Axis Scales and Ticks
## Scales

```python
fig, axs = plt.subplots(1, 2, figsize=(5, 2.7), layout="constrained")
xdata = np.arange(len(data1))  # make an ordinal for this
data = 10**data1
axs[0].plot(xdata, data)

axs[1].set_yscale("log")
axs[1].plot(xdata, data)
```

## Tick Locators and Formatters

```python
fig, axs = plt.subplots(2, 1, layout="constrained")
axs[0].plot(xdata, data1)
axs[0].set_title("Automatic ticks")

axs[1].plot(xdata, data1)
axs[1].set_xticks(np.arange(0, 100, 30), ["zero", "30", "sixty", 90])
axs[1].set_yticks([-1.5, 0, 1.5])
axs[1].set_title("Manual ticks")
```

## Plotting Dates and Strings

```python
from matplotlib.dates import ConciseDateFormatter

fig, ax = plt.subplots(figsize=(5, 2.7), layout="constrained")
dates = np.arange(np.datetime64("2021-11-15"), np.datetime64("2021-12-25"), np.timedelta64(1, "h"))
data = np.cumsum(np.random.randn(len(dates)))
ax.plot(dates, data)
ax.xaxis.set_major_formatter(ConciseDateFormatter(ax.xaxis.get_major_locator()))

plt.show()
```

```python
# Plotting categorical values
fig, ax = plt.subplots(figsize=(5, 2.7), layout="constrained")
categories = ["turnips", "rutabaga", "cucumber", "pumpkins"]

ax.bar(categories, np.random.rand(len(categories)))
```

## Additional Axis Objects
Use <b style="color:#FFFF00;">twinx</b> to add a new Axes with an invisible x-axis and a y-axis positioned to the right

```python
fig, (ax1, ax3) = plt.subplots(1, 2, figsize=(7, 2.7), layout="constrained")
(l1,) = ax1.plot(t, s)
ax2 = ax1.twinx()
(l2,) = ax2.plot(t, range(len(t)), "C1")
ax2.legend([l1, l2], ["Sine (left)", "Straight (right)"])

ax3.plot(t, s)
ax3.set_xlabel("Angle [rad]")
ax4 = ax3.secondary_xaxis("top", (np.rad2deg, np.deg2rad))
ax4.set_xlabel("Angle [o]")
```

# Color Mapped Data

```python
from matplotlib.colors import LogNorm

X, Y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
Z = (1 - X / 2 + X**5 + Y**3) * np.exp(-(X**2) - Y**2)

fig, axs = plt.subplots(2, 2, layout="constrained")
pc = axs[0, 0].pcolormesh(X, Y, Z, vmin=-1, vmax=1, cmap="RdBu_r")
fig.colorbar(pc, ax=axs[0, 0])
axs[0, 0].set_title("pcolormesh()")

co = axs[0, 1].contourf(X, Y, Z, levels=np.linspace(-1.25, 1.25, 11))
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title("contourf()")

pc = axs[1, 0].imshow(Z**2 * 100, cmap="plasma", norm=LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc, ax=axs[1, 0], extend="both")
axs[1, 0].set_title("imshow() with LogNorm()")

pc = axs[1, 1].scatter(data1, data2, c=data3, cmap="RdBu_r")
fig.colorbar(pc, ax=axs[1, 1], extend="both")
axs[1, 1].set_title("scatter()")
```
