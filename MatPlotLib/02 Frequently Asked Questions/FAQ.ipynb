```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
```

#### 1. Too many or out of order ticks in the x-axis

```python
fig, ax = plt.subplots(1, 2, figsize=(8, 2), layout="constrained")

ax[0].set_title("Ticks seem out of order/misplaced", color="red", fontsize=14)
x = ["5", "20", "1", "9"]  # NOTE: these are strings!
y = [5, 20, 1, 9]  #      these are numbers
# The data passed in x, are strings not numbers so matplotlib
# reads them as categorical variables and puts one tick per category!
ax[0].plot(x, y, "d")
ax[0].tick_params(axis="x", labelcolor="red", labelsize=15)

# As above x is a list of strings and so it is treated as categorical variables:
#   matplotlib places on tick per category
#   plots them in the order they appear
ax[1].set_title("Many ticks", color="brown", fontsize=14)
x = [str(xx) for xx in np.arange(100)]  # strings
y = np.arange(100)
ax[1].plot(x, y)
ax[1].tick_params(axis="x", labelcolor="brown", labelsize=15)
```

#### 3. Check wheter a Figure is empty

```python
# A Figure is considered to be empty if it does not contain
# any Artists but its background patch.
def is_empty(figure: plt.figure):
    """
    Return whether the figure contains no Artists (other than the default background patch.)
    """
    contained_artists = figure.get_children()
    return len(contained_artists) <= 1

is_empty(fig)
```

#### 4. Find all objects in a figure of a certain type

```python
def myfunc(x):
    return hasattr(x, "set_color")


for o in fig.findobj(myfunc):
    o.set_color("blue")
```
